```python
import os
import pandas as pd
import numpy as np
```

```python
ratings_data=pd.read_csv('ratings.csv')
ratings_data.head()
```

```python
movie_names=pd.read_csv('movies.csv')
movie_names.head()
```

```python
movie_data=pd.merge(ratings_data,movie_names,on='movieId')
```

```python
movie_data.head()
```

```python
movie_data.groupby('title')['rating'].mean().head()
```

```python
movie_data.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
movie_data.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
rating_mean_count=pd.DataFrame(movie_data.groupby('title')['rating'].mean())
```

```python
rating_mean_count['rating_counts']=pd.DataFrame(movie_data.groupby('title')['rating'].count())
```

```python
rating_mean_count.sort_values(ascending=False,by='rating')
```

```python
#Let's see if there are movies which have more than 4 rating and more than 100 people have rated. 
best_rating=rating_mean_count[rating_mean_count['rating']>4]
```

```python
best_rating[best_rating['rating_counts']>200].sort_values(ascending=False,by='rating_counts')
```
